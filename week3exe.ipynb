### Generate the k-mer Composition of a String

https://rosalind.info/problems/ba3a/

```python
def StringComposition(k, Text):
    return [Text[i:i+k] for i in range(len(Text) - k + 1)]
```

```python
with open ('data/rosalind_ba3a.txt') as f:
    k = int(f.readline().strip())
    Text = f.readline().strip()
    result = StringComposition(k, Text)
    print("\n".join(result))
```

### Reconstruct a String from its Genome Path

https://rosalind.info/problems/ba3b/

```python
def StringSpelledByAGenomePath(Patterns):
    return Patterns[0] + ''.join(pattern[-1] for pattern in Patterns[1:])
```

```python
with open("data/rosalind_ba3b.txt") as f:
    Patterns = [line.strip() for line in f.readlines()]
    result = StringSpelledByAGenomePath(Patterns)
    print(result)
```

### Construct the Overlap Graph of a Collection of k-mers

https://rosalind.info/problems/ba3c/

```python
def OverlapGraph(Patterns):
    adjacency_list = []
    k = len(Patterns[0])
    prefix_map = {pattern[:-1]: pattern for pattern in Patterns}
    
    for pattern in Patterns:
        suffix = pattern[1:]
        if suffix in prefix_map:
            adjacency_list.append(f"{pattern} -> {prefix_map[suffix]}")
    
    return adjacency_list
```

```python
with open("data/rosalind_ba3c.txt") as f:
    Patterns = [line.strip() for line in f.readlines()]
    result = OverlapGraph(Patterns)
    print("\n".join(result))
```
